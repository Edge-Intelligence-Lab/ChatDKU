```python
import pickle
# 加载原documents
pkl_file_path = "/home/Glitterccc/projects/DKU_LLM/RAG_data/documents.pkl"
with open(pkl_file_path, 'rb') as f:
    pre_documents = pickle.load(f)
```

```python
file_types = []
for doc in pre_documents:
    if doc.metadata["file_type"] not in file_types:
        file_types.append(doc.metadata["file_type"])
```

```python
file_types
```

```python
documents_pdf = []
for doc in pre_documents:
    if doc.metadata["file_type"] == 'application/pdf':
        documents_pdf.append(doc)
```

```python
# 保存html_documents
with open('old_pdf_documents.pkl', 'wb') as f:
    pickle.dump(documents_pdf, f)
```
