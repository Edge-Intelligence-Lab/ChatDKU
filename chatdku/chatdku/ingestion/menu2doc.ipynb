```python
import pandas as pd
from openpyxl import load_workbook

def fill_merged_cells(sheet):
    merged_ranges = list(sheet.merged_cells.ranges)
    for merged_cell in merged_ranges:
        merged_cell_range = merged_cell.coord
        # 获取合并单元格的起始和结束行列
        min_col, min_row, max_col, max_row = merged_cell.min_col, merged_cell.min_row, merged_cell.max_col, merged_cell.max_row
        # 获取合并单元格的值
        value = sheet.cell(row=min_row, column=min_col).value
        print(value)
        # 解除合并单元格
        sheet.unmerge_cells(merged_cell_range)
        # 填充合并单元格中的每一个单元格
        for row in range(min_row, max_row + 1):
            for col in range(min_col, max_col + 1):
                sheet.cell(row=row, column=col,value = value)
        # 重新合并单元格
        sheet.merge_cells(merged_cell_range)

def read_excel_with_merged_cells(file_path, sheet_name=None):
    # 加载工作簿
    wb = load_workbook(file_path)
    # 获取指定工作表
    if sheet_name:
        sheet = wb[sheet_name]
    else:
        sheet = wb.active
    # 处理合并单元格
    fill_merged_cells(sheet)
    # 将工作表转换为 DataFrame
    data = sheet.values
    columns = next(data)[0:]  # 获取列名
    df = pd.DataFrame(data, columns=columns)
    return df
```

```python

# 示例用法
file_path = '0902—0908_week_menu_.xlsx'
df = read_excel_with_merged_cells(file_path, sheet_name='自选称重餐线')  # 如果不指定 sheet_name，默认读取第一个工作表

# 打印 DataFrame 以验证读取是否成功
print(df)
```
