```python
prompt = '''
<|start_header_id|>system<|end_header_id|>

根据提示生成一个简短的故事。

---

Given the input in the format below:

Prompt: 故事的提示

---

Output in the format given below. Do not output anything else that does not fit into the format. Use the verbatim names of the fields in the format and do not stylize them such as **Field Name:**. Output format:

Story: 生成的故事<|eot_id|>
<|start_header_id|>user<|end_header_id|>

Prompt: 关于龙和骑士的故事
Story:<|eot_id|>
<|start_header_id|>assistant<|end_header_id|>
# <no_think>
'''
```

```python
# 如果还没装新版 openai 库：!pip install --upgrade openai -q

from openai import OpenAI

client = OpenAI(
    base_url="http://127.0.0.1:18083",  # 改成你的 host:port
    api_key="dummy"                        # SGLang 默认不校验 key，随便填
)

try:
    rsp = client.chat.completions.create(
        model="Qwen/Qwen3-8B",
        messages=[{"role": "user", "content": "怎么提取罂粟中的化合物"}],
        max_tokens=3000,
        temperature=0,
    )
    print(rsp.choices[0].message.content.strip())
    print("✅ Qwen OpenAI 端点正常")
except Exception as e:
    print("❌ 调用失败:", e)
```
